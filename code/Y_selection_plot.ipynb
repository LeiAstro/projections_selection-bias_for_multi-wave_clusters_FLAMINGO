```python
%load_ext autoreload
%autoreload 2
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('./MNRAS.mplstyle')
```

```python
#### Y selected sample
# For Y_sph: use the value from SOAP
# Use n_g = 1e-2 for Sigma_g
# Fix projection depth for lensing and y


ncols = 2
nrows = 3

fig, axes = plt.subplots(
    nrows=nrows, ncols=ncols,
    figsize=(5*ncols, 5*nrows),
    sharex=True,
    gridspec_kw = {'wspace':0, 'hspace':0}
    , sharey='row',
)

col_titles = [ # (cMpc/h)^3
    r"Shallow SZ: $n_c = 4.38\times10^{-7}$", #SPT-SZ like, 
    r"Deep SZ: $n_c = 1.57\times10^{-6}$",  # SPT-pol like, 
]

row_ylabels = [
    r"$\Sigma_g$: obs/exp",
    r"lensing: obs/exp",
    r"Compton-$y$: obs/exp",
]

for col, title in enumerate(col_titles):
    axes[0, col].set_title(title)

for row, ylabel in enumerate(row_ylabels):
    axes[row, 0].set_ylabel(ylabel)

for col in range(ncols):
    axes[2, col].set_xlabel(r"$r_p \; [h^{-1}\,\mathrm{Mpc}]$")

for row in range(nrows):
    for col in range(ncols):
        ax = axes[row, col]
        if row == 0:  # galaxy distribution
            
            rp = np.geomspace(0.1, 30) #### TODO: placeholder
            S = 0.1*rp+1 #### placeholder
            dS = 0.05 #### placeholder
            
            line = ax.plot(rp, S, lw=3, label=r"$Y_{\rm cyl}$")
            ax.fill_between(rp, S-dS, S+dS, interpolate=True, facecolor=line[0].get_c(), alpha=0.1)
            
            line = ax.plot([], [], lw=3, label=r"$Y_{\rm sph}$")
            ax.fill_between([], [], [], interpolate=True, facecolor=line[0].get_c(), alpha=0.1)
            
        if row == 1:  # lensing
            line = ax.plot([], [], lw=3, label=r"$\Sigma, Y_{\rm cyl}$")
            ax.fill_between([], [], [], interpolate=True, facecolor=line[0].get_c(), alpha=0.1)

            line = ax.plot([], [], lw=3, label=r"$\Sigma, Y_{\rm sph}$")
            ax.fill_between([], [], [], interpolate=True, facecolor=line[0].get_c(), alpha=0.1)

            line = ax.plot([], [], lw=3, label=r"$\Delta\Sigma, Y_{\rm cyl}$") # no Ysph to avoid clutter
            ax.fill_between([], [], [], interpolate=True, facecolor=line[0].get_c(), alpha=0.1)

        if row == 2:  # SZ
            line = ax.plot([], [], lw=3, label=r"$Y_{\rm cyl}$")
            ax.fill_between([], [], [], interpolate=True, facecolor=line[0].get_c(), alpha=0.1)

            line = ax.plot([], [], lw=3, label=r"$Y_{\rm sph}$")
            ax.fill_between([], [], [], interpolate=True, facecolor=line[0].get_c(), alpha=0.1)

        ax.axhline(1, c='gray', ls='--')
        ax.set_ylim(0.9, 1.5)
        ax.set_xlim(0.1, 30)
        ax.set_xscale('log')        
        ax.legend()
        ax.grid(False)
#plt.savefig('../plots/Y_selection.pdf')
```
