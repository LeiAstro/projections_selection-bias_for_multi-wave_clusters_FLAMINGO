```python
%load_ext autoreload
%autoreload 2
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('./MNRAS.mplstyle')

import os, glob
import h5py
```

```python
fig, axes = plt.subplots(
    2, 1,
    sharex=True,
    figsize=(6, 8),
    gridspec_kw={'hspace': 0}
)

# loop through 4 mass-ranked abundance bins

nbin = 4 
for ibin in range(nbin):

    # ----------- TODO: placeholder  ----------- 
    rp = np.geomspace(0.1, 30)
    Sigma = 0*rp+1.05 
    DeltaSigma = Sigma 
    err = 0*rp+0.01
    # ---------------------- 

    # top panel for Sigma
    line = axes[0].plot(rp, Sigma, label=f'bin {ibin}')
    axes[0].fill_between(rp, Sigma-err, Sigma+err, interpolate=True, facecolor=line[0].get_c(), alpha=0.3)
    
    # bottom panel for DeltaSigma
    line = axes[1].plot(rp, DeltaSigma)
    axes[1].fill_between(rp, DeltaSigma-err, DeltaSigma+err, interpolate=True, facecolor=line[0].get_c(), alpha=0.3)


axes[0].set_ylabel(r'$\Sigma$: hydro/DMO')
axes[1].set_ylabel(r'$\Delta\Sigma$: hydro/DMO')


axes[0].tick_params(labelbottom=False)
axes[1].set_xlabel(r'$r_p$')
axes[0].legend()

for ipanel in [0, 1]:
    ax = axes[ipanel]
    ax.axhline(1, c='gray', ls='--')
    ax.set_ylim(0.85, 1.15)
    ax.set_xlim(0.1, 30)
    ax.set_xscale('log')        
    ax.grid(False)
plt.show()
```
